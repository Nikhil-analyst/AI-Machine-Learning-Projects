# Let's create a Pivot Table now

```python
book_pivot = final_rating.pivot_table(columns = "user_id",index= "title", values = "rating")
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[237, :].values.reshape(1,-1)
```

```python
distnaces , suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1,-1), n_neighbors = 6)
```

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index==book_name)[0][0]
    distnaces , suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1,-1), n_neighbors = 6)
    
    for i in range(len(suggestions[0])):
        if i == 0:
            print(color.BOLD + "If you like: " + color.END, book_pivot.index[suggestions[0][0]])
            print(color.BOLD + "You will also like: "+ color.END,end = "")
        else:
            print(book_pivot.index[suggestions[0][i]], end = ", ")
```

```python
recommend_book("1984")
```
